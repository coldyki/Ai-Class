```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow import keras

df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/wine.csv")
df
```

```python
y=df['Wine']
y
```

```python
X = df.drop("Wine", axis=1)
X
```

```python
from sklearn.model_selection import train_test_split

X1_train,X1_test, y1_train,y1_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.ensemble import RandomForestClassifier
import warnings

clf_lr = LogisticRegression(random_state=0)
clf_lr.fit(X1_train, y1_train)

pred_lr = clf_lr.predict(X1_test)

print ("\n--- Logistic Regression Classifier ---")
print (accuracy_score(y1_test, pred_lr))
print (confusion_matrix(y1_test, pred_lr))
```

```python
clf_dt1 = DecisionTreeClassifier(random_state=0)
clf_dt1.fit(X1_train, y1_train)

pred_dt = clf_dt1.predict(X1_test)

print ("\n--- Decision Tree Classifier ---")
print (accuracy_score(y1_test, pred_dt))
print (confusion_matrix(y1_test, pred_dt))
```

```python
print ("\n--- Radom Forest ---")
rf_clf1 = RandomForestClassifier(random_state=0)
rf_clf1.fit(X1_train, y1_train)
pred = rf_clf1.predict(X1_test)
print(accuracy_score(y1_test,pred))
print (confusion_matrix(y1_test, pred))
```

```python
df2 = df.drop('Wine', axis=1)
df2
```

```python
y2 = df2['Alcohol']
X2 = df2.drop('Alcohol', axis =1)

X2
```

```python
X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y2,test_size=0.25,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

model = LinearRegression()
model.fit(X2_train, y2_train)
ly_preds = model.predict(X2_test)

print('평균제곱근오차', mean_squared_error(ly_preds, y2_test))

def mse_np(actual, predicted):
    return np.mean((np.array(actual) - np.array(predicted)) ** 2)

print('평균제곱근오차', mse_np(ly_preds, y2_test))

def mse(actual, predicted):
    sum_square_error = sum((a - p) ** 2 for a, p in zip(actual, predicted))
    mean_square_error = sum_square_error / len(actual)
    return mean_square_error

print('평균제곱근오차', mse(ly_preds, y2_test))
```

```python
from sklearn.tree import DecisionTreeRegressor

model = DecisionTreeRegressor()
model.fit(X2_train, y2_train)

from sklearn.metrics import mean_squared_error

dy_preds = model.predict(X2_test)
print('평균제곱근오차', mean_squared_error(dy_preds, y2_test))
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor()
model.fit(X2_train, y2_train)

from sklearn.metrics import mean_squared_error

ry_preds = model.predict(X2_test)
print('평균제곱근오차', mean_squared_error(ry_preds, y2_test))
```

```python
from tensorflow.keras import layers, models

Y = pd.get_dummies(y).values
X = X.values
```

```python
X_train,X_test, y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
model = models.Sequential([
    layers.Dense(64, activation="relu", input_shape=(X.shape[1],)),
    layers.Dropout(0.3),
    layers.Dense(32, activation="relu"),
    layers.Dense(Y.shape[1], activation="softmax")   # 클래스 수 맞춤
])

model.compile(
    optimizer="adam",
    loss="categorical_crossentropy",   # one-hot이면 categorical_crossentropy
    metrics=["accuracy"]
)

history = model.fit(
    X_train, y_train,
    validation_split=0.2,
    epochs=50,
    batch_size=16,
    verbose=1
)
```

```python
y_pred = model.predict(X_test)

y_test_class = np.argmax(y_test, axis=1)
y_pred_class = np.argmax(y_pred, axis=1)

# 학습 곡선
loss = history.history['loss']
val_loss = history.history['val_loss']
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']


y_pred = model.predict(X_test)

y_test_class = np.argmax(y_test,axis=1)
y_pred_class = np.argmax(y_pred,axis=1)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

loss = history.history['loss']
val_loss= history.history['val_loss']
epochs = range(1, len(loss) + 1)
plt.plot(epochs, loss, 'y', label='Training loss')
plt.plot(epochs, val_loss, 'r', label='Validation loss')
plt.title('Training and validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()

acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
plt.plot(epochs, acc, 'y', label='Training acc')
plt.plot(epochs, val_acc, 'r', label='Validation acc')
plt.title('Training and validation accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test_class,y_pred_class))
print(confusion_matrix(y_test_class,y_pred_class))
```
